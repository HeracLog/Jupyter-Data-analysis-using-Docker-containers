```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./books.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.duplicated().sum()
```

```python
condition = df['original_title'].str.contains('Harry Potter') 
rdf = df[condition]
rdf.head()
```

```python
rdf = rdf.drop([
    'book_id','best_book_id', 'goodreads_book_id','work_id','isbn','isbn13','authors',
    'title','language_code','image_url','small_image_url'
    ],axis=1).drop(1036)
rdf.head()
```

```python
plt.figure(figsize=(14, 7), dpi=70)
plt.subplot(121)
sns.barplot(data=rdf, y='original_title', x='average_rating')
plt.xlim(4, 5)
plt.title('Average ratings')
plt.subplot(122)
sns.barplot(data=rdf, y='original_title', x='work_ratings_count')
plt.title('Work Ratings Count')
plt.tight_layout()
plt.savefig('ratings.png')
plt.show()
```

```python
plt.figure(figsize=(20,10),dpi=80)
plt.subplot(121)
plt.pie(rdf['ratings_5'],autopct='%1.1f%%')
plt.legend(title='5-star ratings',loc="best",labels= rdf['original_title'])
plt.subplot(122)
plt.pie(rdf['books_count'],autopct='%1.1f%%')
plt.legend(title='Books count',loc="upper left",labels= rdf['original_title'])
plt.savefig('piecharts.png')
plt.show()
```

```python
sdf = rdf.describe()
sdf
```

```python
sdf.drop(['count'],inplace=True)
sdf
```

```python
rsdf = sdf.drop(['original_publication_year','average_rating','ratings_count','work_ratings_count','work_text_reviews_count','books_count'],axis=1).transpose()
ratings = list(rsdf.index)
mean = list(rsdf['mean'])
pltdf = pd.DataFrame({'mean':mean,'ratings':ratings})
pltdf
```

```python
plt.figure(figsize=(14,7),dpi=100)
sns.barplot(pltdf, x='mean',y='ratings')
plt.title("Average ratings for all books")
plt.savefig('ratingsAvg.png')
plt.show()
```
